Loading data into google colab 

```python
#Loading data into google colab 

overall = pd.read_excel("/content/Workshop Data.xlsx", sheet_name="Overall")
Project_details = pd.read_excel("/content/Workshop Data.xlsx", sheet_name="Project Detail")
employee_details = pd.read_excel("/content/Workshop Data.xlsx", sheet_name="Employee Detail ")
```

Viewing the first few rows of our dataset 

```python
# Let us view the first couple rows of our dataset to get a feel 

overall.head(2)
```

Let us join our dataset first 

```python
df_1 = pd.merge(Project_details, overall)
df_2 = pd.merge(employee_details, df_1)
df_2
```

```python
df_2['Project Name '].unique()
```

```python
#let us view how many rows and columns we have in the dataset 
df_2.shape
```

```python
#let understand a little more 

df_2.describe()
```

```python
#Understanding the dtypes in our dataframe 

df_2.dtypes
```

Let us add a new column in the dataset 

```python
# new column for total hours worked on a project

df_2['total_hours_worked'] = df_2['Hours Worked Per Day '] * df_2['Number of days on project ']
df_2.head()
```

```python
# Let try to find out the revenue generated per hour of work 

df_2['revenue_per_hour'] = round(df_2['Revenue Generated '] / df_2['total_hours_worked'],2)
df_2
```

```python
#plot nan values to visualised 
df_2.info()
```

```python
#Using seaborn library to plot missing data

sns.heatmap(df_2.isnull(), cbar=True)
```

```python
# Let us now drop null values from our dataset 
df_3 = df_2.dropna()
```

```python
df_3
```

```python
df_3.columns
```

```python
# Let us group the data to bette understand how each project did: 

table_1 = pd.pivot_table(df_3, values='Revenue Generated '  , index=['Employee Name '],aggfunc=np.sum)
table_1
```

```python
table_2 = pd.pivot_table(df_3, values=['total_hours_worked','revenue_per_hour'] , index=['Project Name '])
table_2.sort_values(by=['revenue_per_hour'], ascending=False)
```

All projects that generated lowered than ▶

* 25,000 in revenue 
* above 50,000 in revenue 
* between 30,000 & 40,000 in revenue 




```python
#Filtering

# Let us find out the total revenue from Janurary to June & July to december 

less_than_20k = df_3.loc[df_3['Revenue Generated '] < 25000]
less_than_20k
```

```python
more_than_50k = df_3.loc[df_3['Revenue Generated '] > 50000]
more_than_50k
```

```python
between_30_to_40 = df_3.loc[(df_3['Revenue Generated '] > 30000) & (df_3['Revenue Generated '] < 40000)]
between_30_to_40
```

```python
correlation = df_3.corr()
correlation
```

```python
# correlation heatmap 
import seaborn as sns 
```

```python
sns.heatmap(correlation);
# Increase the size of the heatmap.
plt.figure(figsize=(20, 15))
```

```python
#visualisation 

fig_dims = (20, 10)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x = 'Project Name ', y = 'Revenue Generated ', ax=ax, data=df_3, palette="Blues_d")
```

```python
#visualisation 2 


fig_dims = (20, 10)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x = 'Employee Name ', y = 'total_hours_worked', ax=ax, data=df_3)

#set labels 
plt.xlabel("Employee Name", size=15)
plt.ylabel("Total Hours Worked", size=15)
plt.title("Total Hours worked per employee", size=18)
plt.tight_layout()
```
